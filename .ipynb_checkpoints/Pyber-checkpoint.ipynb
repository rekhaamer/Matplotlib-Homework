Trend 1:Rural areas have lower number of rides while customers pay higher fares
Trend 2:Urban areas have higher number of rides while customers pay relatively low fares
Trend 3: Co-relating all pie charts, a higher proportion of drivers in Urban areas doesn't result in equal proporting of rides and fare. 

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as se
city_file=os.path.join('city_data.csv')
city_file_df=pd.read_csv(city_file)
city_file_df.head()
```

```python
ride_file=os.path.join('ride_data.csv')
ride_file_df=pd.read_csv(ride_file)
ride_file_df.head()
```

```python
JoinedTable=pd.merge(ride_file_df, city_file_df, on="city",how="left")
JoinedTable.head()
```

```python
urban = JoinedTable[JoinedTable['type'] == 'Urban']
urbanRideCount=urban.groupby('city').count()['ride_id']
urbanAverageFare=urban.groupby('city').mean()['fare']
urbanDriverCount=urban.groupby('city').mean()['driver_count']
urbanPlot=pd.DataFrame({'Average Fare':urbanAverageFare,'Ride Count':urbanRideCount, 
                        'Driver Count':urbanDriverCount})
urban.head()
```

```python
suburban = JoinedTable[JoinedTable['type'] == 'Suburban']
suburbanRideCount=suburban.groupby('city').count()['ride_id']
suburbanAverageFare=suburban.groupby('city').mean()['fare']
suburbanDriverCount=suburban.groupby('city').mean()['driver_count']
suburbanPlot=pd.DataFrame({'Average Fare':suburbanAverageFare,'Ride Count':suburbanRideCount, 
                        'Driver Count':suburbanDriverCount})
suburbanPlot.head()
```

```python
rural = JoinedTable[JoinedTable['type'] == 'Rural']
ruralRideCount=rural.groupby('city').count()['ride_id']
ruralAverageFare=rural.groupby('city').mean()['fare']
ruralDriverCount=rural.groupby('city').mean()['driver_count']
ruralPlot=pd.DataFrame({'Average Fare':ruralAverageFare,'Ride Count':ruralRideCount, 
                        'Driver Count':ruralDriverCount})
ruralPlot.head()
```

```python
plt.scatter(urbanRideCount, urbanAverageFare,s=(10*urbanDriverCount),marker='o',c='lightcoral', edgecolors="black",
            label="Urban", alpha=0.8,linewidths=1 )
plt.scatter(suburbanRideCount, suburbanAverageFare,s=(10*suburbanDriverCount),marker='o',c='lightblue', edgecolors="black",
            label="Suburban", alpha=0.8 )
plt.scatter(ruralRideCount,ruralAverageFare,s=(10*ruralDriverCount),marker='o',c='gold', edgecolors="black",
           label="Rural", alpha=0.8)
plt.ylim(15, 55)
plt.xlim(0, 40)
plt.title('Pyber Ride Sharing Data (2016)')
plt.ylabel('Average Fare($)')
plt.xlabel('Total Number of Rides (Per City)')
lgnd=plt.legend(loc="upper right", title="City Types", scatterpoints=1)
lgnd.legendHandles[0]._sizes = [10]
lgnd.legendHandles[1]._sizes = [10]
lgnd.legendHandles[2]._sizes = [10]
plt.text (4,2, "Note: Size of bubble indicates number of drivers in city")
plt.grid()
plt.savefig("pyber_analysis.png")
plt.show()
```

```python
CityCenterFare=JoinedTable.groupby('type').sum()
CityCenterFare
```

```python
plt.pie(CityCenterFare['fare'],explode=(0, 0, 0.1),autopct="{:.1f}%".format, labels=CityCenterFare.index, 
        colors=["Gold", "lightskyblue", "lightcoral"],startangle=150,shadow=True)
plt.axis('equal')
plt.title("% of Total Fares by City Type")
plt.show()
```

```python
CityCenterRide=JoinedTable.groupby('type').count()
CityCenterRide
```

```python
plt.pie(CityCenterRide['ride_id'],explode=(0, 0, 0.1),autopct="{:.1f}%".format, labels=CityCenterRide.index, 
        colors=["Gold", "lightskyblue", "lightcoral"],startangle=150,shadow=True)
plt.axis('equal')
plt.title("% of Total Rides by City Type")
plt.show()
```

```python
CityCenterDriver=JoinedTable.groupby(['city']).max()
CityCenterDriver['type']=JoinedTable.groupby('city')['type'].max()
CityCenterDriver=CityCenterDriver.groupby('type').sum()
CityCenterDriver.head()
```

```python
plt.pie(CityCenterDriver['driver_count'],explode=(0, 0, 0.1),autopct="{:.1f}%".format, labels=CityCenterDriver.index, 
        colors=["Gold", "lightskyblue", "lightcoral"],startangle=150, shadow=True)
plt.axis('equal')
plt.title("% of Total Drivers by City Type")
plt.show()
```
